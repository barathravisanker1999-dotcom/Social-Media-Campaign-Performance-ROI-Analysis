```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
```

```python
df['Date'].head()
```

```python
plt.hist(df['Clicks'], bins=1000)
plt.title("Distribution of Clicks")
plt.xlabel("Clicks")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.countplot(x='Channel_Used', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(x='Impressions', y='Clicks', data=df)
plt.show()
```

```python
sns.boxplot(x='Channel_Used', y='ROI', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
```

```python
df.groupby('Month')['Clicks'].sum().plot(kind='line')
plt.title("Monthly Click Trend")
plt.show()
```
